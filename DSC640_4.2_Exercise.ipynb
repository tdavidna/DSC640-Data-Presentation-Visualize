```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt, cm, colors
import seaborn as sns
```

```python
# Reading the excel file into a DataFrame

home_dir = 'C:/David/Study_MS/DSC640/Data'
crimerate_df = pd.read_csv(home_dir + '/crimeRatesByState.csv')
```

### Scatter Plot

```python
plt.figure(figsize=(18,6))
murder = plt.scatter(x = crimerate_df['state'], y = crimerate_df['murder'])
forcible_rape = plt.scatter(x = crimerate_df['state'], y = crimerate_df['forcible_rape'])
robbery = plt.scatter(x = crimerate_df['state'], y = crimerate_df['robbery'])
aggravated_assault = plt.scatter(x = crimerate_df['state'], y = crimerate_df['aggravated_assault'])
burglary = plt.scatter(x = crimerate_df['state'], y = crimerate_df['burglary'])
plt.legend((murder, forcible_rape,robbery, aggravated_assault, burglary),
           ('Murder', 'Forcible_Rape', 'Robbery', 'Aggravated_Assault', 'Burglary'),
           scatterpoints=1,
           loc='upper right',
           ncol=3,
           fontsize=12)
plt.title('Python Scatter Plot', color='black', fontweight = 'bold', fontsize = 12)
plt.xticks(rotation=60)
plt.ylabel('Crime Count')
plt.show()
```

### Bubble Chart

```python
plt.figure(figsize=(18,6))
plt.scatter(x = crimerate_df['burglary'], y = crimerate_df['murder'], s=crimerate_df['murder']*100,
            c=crimerate_df['murder'], cmap='viridis')
plt.title('Python Bubble Chart', color='black', fontweight = 'bold', fontsize = 12)
plt.xticks(rotation=60)
plt.show()
```

### Density Map

```python
sns.kdeplot(crimerate_df['murder'], color="blue", shade=True)
plt.title('Python Density Plot for Murder', color='black', fontweight = 'bold', fontsize = 12)
plt.show()
```
